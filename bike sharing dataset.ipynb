```python
def uniplot(col):
    plt.figure()
    plt.subplot(2,2,1)
    sns.histplot(x=hour[col],bins=20,kde=True)
    plt.axvline(x=hour[col].median(),color='red',ymin=0,ymax=1)
    plt.axvline(x=hour[col].mean(),color='green',ymin=0,ymax=1)
    plt.title(f"Histogram of {col}")
    plt.subplot(2,2,2)
    sns.boxplot(x=hour[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
for col in hour.select_dtypes(include=['int','float']).columns:
    uniplot(col)
```
